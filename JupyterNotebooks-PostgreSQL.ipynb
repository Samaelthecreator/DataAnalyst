```python
#Configuramos la conexión a PostgreSQL

from sqlalchemy import create_engine

#detalles de conexión, utilizaremos los mismos datos para la importación.

user = 'postgres'
password = '*********'        #Es importante tener bien los datos, en dado caso que no arrojara un error tipo: OperationalError: (psycopg2.OperationalError) (Background on this error at: https://sqlalche.me/e/14/e3q8)"
host = 'localhost'
port = '5434'
database = 'SITEC, soluciones y servicios'

#Creamos la cadena de conexión.

conn_str = f'postgresql+psycopg2://{user}:{password}@{host}:{port}/{database}'

#Creamos el motor de conexión.

engine = create_engine(conn_str)

query = 'SELECT * FROM compras'

df = pd.read_sql(query,engine)6


df
```
